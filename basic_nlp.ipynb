### Basic NLP Tasks

```python

import spacy
import pandas as pd

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

##### 1. Stop word removal

```python


# Sample text
text = "This is an example sentence to learn natrual language processing."

# Process text
doc = nlp(text)

# Collect token attributes
token_data = []
for token in doc:
    token_data.append({
        "Text": token.text,
        "Lemma": token.lemma_,
        "POS": token.pos_,
        "Tag": token.tag_,
        "Dep": token.dep_,
        "Is_Alpha": token.is_alpha,
        "Is_Stop": token.is_stop,
        "Is_Punct": token.is_punct,
        "Index": token.idx
    })

# Create a DataFrame for structured view
df = pd.DataFrame(token_data)

print("\n Token Details:\n")
print(df.to_string(index=False))

```

```python
def remove_stop_word(text):
    ''' Remove the stop words.
    Returns non stopwords'''
    doc = nlp(text)
    filtered_tokens = [token.text for token in doc if not token.is_stop]
    return filtered_tokens
```

#### 2. Lemmatization

```python


def lemmatize_text(text):
    """
    Lemmatize the input text using spaCy.
    Returns a list of lemmas.
    """
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc if not token.is_punct]
    return lemmas

# Example
text = "The striped bats are hanging on their feet for best"
print("Lemmatized:", lemmatize_text(text))
```
